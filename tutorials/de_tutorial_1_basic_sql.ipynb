```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('orders.db')

!pip -q install --upgrade ipython
!pip -q install --upgrade ipython-sql


%load_ext sql
%sql sqlite:///orders.db

%config SqlMagic.autopandas = True
%config SqlMagic.feedback = False
%config SqlMagic.displaycon = False
```

```python
# prompt: create a table called orders with an orderid and custid column

%sql CREATE TABLE orders (orderid INTEGER PRIMARY KEY, custid INTEGER);
%sql CREATE TABLE orderdetail (orderdetailid INTEGER PRIMARY KEY, orderid INTEGER);
```

```python
%sql INSERT INTO orders VALUES (25, 1);
%sql INSERT INTO orders VALUES (30, 2);
%sql INSERT INTO orders VALUES (15, 3);
%sql INSERT INTO orders VALUES (28, 4);

%sql INSERT INTO orderdetail VALUES (1, 25);
%sql INSERT INTO orderdetail VALUES (2, 25);
%sql INSERT INTO orderdetail VALUES (3, 15);
%sql INSERT INTO orderdetail VALUES (4, 15);
```

```python
pd.read_sql_query("SELECT * FROM orders ;", con)
```

```python
pd.read_sql_query("SELECT * FROM orderdetail;", con)
```

```python
pd.read_sql_query("SELECT * FROM orders o left join orderdetail d on o.orderid = d.orderid;", con)
```

```python
pd.read_sql_query("SELECT * FROM orders o inner join orderdetail d on o.orderid = d.orderid;", con)
```
